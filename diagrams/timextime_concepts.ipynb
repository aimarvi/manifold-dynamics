```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
rng = np.random.default_rng(42)

def make_rdm(base_similarity, noise_sd=0.1):
    """
    base_similarity: square numpy array with values in [0, 1]
    Returns an RDM where larger values mean more dissimilar.
    """
    noise = rng.normal(0, noise_sd, base_similarity.shape)
    sim = base_similarity + noise

    # keep diagonal maximally similar
    np.fill_diagonal(sim, 1.0)

    # symmetrize and clip
    sim = (sim + sim.T) / 2
    sim = np.clip(sim, 0, 1)

    # convert similarity -> dissimilarity
    rdm = 1 - sim

    # diagonal of an RDM should be 0
    np.fill_diagonal(rdm, 0.0)
    return rdm

n = 35

start1, end1 = 5, 30

start21, end21 = 5, 17
start22, end22 = 17, 30


# 1) One large similarity block from indices 2..6
sim1 = np.zeros((n, n))
np.fill_diagonal(sim1, 1.0)
sim1[start1:end1, start1:end1] = 0.6
np.fill_diagonal(sim1, 1.0)
rdm1 = make_rdm(sim1)

# 2) Two similarity blocks: 2..3 and 4..6
sim2 = np.zeros((n, n))
np.fill_diagonal(sim2, 1.0)
sim2[start21:end21, start21:end21] = 0.6
sim2[start22:end22, start22:end22] = 0.6
np.fill_diagonal(sim2, 1.0)
rdm2 = make_rdm(sim2)

# 3) Only on-diagonal similarity
sim3 = np.eye(n)
rdm3 = make_rdm(sim3)

# plot
titles = [
    "one-block",
    "two-block",
    "identity"
]

for rdm, title in zip([rdm1, rdm2, rdm3], titles):
    plt.figure(figsize=(4, 4))
    sns.heatmap(rdm, vmin=0, vmax=1, square=True, cbar=False)
    # plt.colorbar(label="Dissimilarity")
    # plt.title(title)
    plt.yticks(np.arange(5, n+5, 5), np.arange(5, n+5, 5))
    plt.xticks(np.arange(5, n+5, 5))
    plt.savefig(f"./{title}.png", transparent=True, dpi=300, bbox_inches="tight")
    plt.show()
```
